```python
plt.figure(figsize=(12, 6))
meal_dish_dist = orders_df.groupby(['Dish Name', 'Meal Type']).size().unstack(fill_value=0)

ax = meal_dish_dist.plot(kind='bar', stacked=True)
plt.title('Distribution of Dishes Across Meal Types')
plt.xlabel('Dish Name')
plt.ylabel('Number of Orders')
plt.legend(title='Meal Type')
plt.xticks(rotation=45, ha='right')

for c in ax.containers:
    ax.bar_label(c, label_type='center')
plt.tight_layout()
plt.savefig("Dist_Meal_type.png", dpi=300, bbox_inches='tight')
plt.show()
print("\nDish Distribution by Meal Type:")
print("-------------------------------")
print(meal_dish_dist.to_string())
print("\nPercentage Distribution:")
print("----------------------")
percentage_dist = meal_dish_dist.div(meal_dish_dist.sum(axis=1), axis=0) * 100
print(percentage_dist.round(1).to_string(), "% within each dish")
```

```python
print("\nKey Insights:")
print("-------------")
print("1. User Demographics:")
print(f"   - Average user age: {user_df['Age'].mean():.1f} years")
print(f"   - Most common favorite meal: {user_df['Favorite Meal'].mode()[0]}")

print("\n2. Order Patterns:")
print(f"   - Most popular dish: {orders_df['Dish Name'].mode()[0]}")
print(f"   - Average order rating: {orders_df['Rating'].mean():.2f}")
print(f"   - Most common order time: {orders_df['Time of Day'].mode()[0]}")

print("\n3. Cooking Sessions:")
print(f"   - Highest rated dish: {sessions_df.loc[sessions_df['Session Rating'].idxmax(), 'Dish Name']}")
print(f"   - Average session duration: {sessions_df['Duration (mins)'].mean():.1f} minutes")
```

```python
dish_metrics = pd.DataFrame()
avg_ratings = orders_df.groupby('Dish Name')['Rating'].mean().round(2)
dish_metrics['Avg Order Rating'] = avg_ratings

avg_duration = sessions_df.groupby('Dish Name')['Duration (mins)'].mean().round(1)
dish_metrics['Avg Session Duration (mins)'] = avg_duration

dish_user_age = orders_df.merge(user_df, on='User ID').groupby('Dish Name')['Age'].mean().round(1)
dish_metrics['Avg Customer Age'] = dish_user_age

order_counts = orders_df.groupby('Dish Name').size()
dish_metrics['Total Orders'] = order_counts

dish_metrics = dish_metrics.sort_values('Total Orders', ascending=False)

print("Comprehensive Dish Analysis")
print("--------------------------")
print(dish_metrics.to_string())
```
